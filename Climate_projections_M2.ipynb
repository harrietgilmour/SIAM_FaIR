# Plot #1

```python
fig = plt.figure(figsize=[16,10])
plt.suptitle('Climate projections of SIAM scenarios',size=16,weight='bold',y=0.93)
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.tick_params(axis='both', which='major', labelsize=13)
ax2.tick_params(axis='both', which='major', labelsize=13)
ax3.tick_params(axis='both', which='major', labelsize=13)
ax4.tick_params(axis='both', which='major', labelsize=13)

#this section is my modified RCP26 scenario with the SIAM D2022 emissions
ax1.plot(rcp26_data_m_D2022.Emissions.year[215:336], rcp26_data_m_D2022.Emissions.co2_fossil[215:336], color='green',label='D2022')
ax2.plot(rcp26_data_m_D2022.Emissions.year[215:336], C26_D2022[:, 0][215:336], color='green',linestyle='dashed')
ax3.plot(rcp26_data_m_D2022.Emissions.year[215:336], np.sum(F26_D2022, axis=1)[215:336], color='green',linestyle='dashed')
ax4.plot(rcp26_data_m_D2022.Emissions.year[215:336], T26_D2022[215:336], color='green',linestyle='dashed')

#this section is my modified RCP26 scenario with the SIAM D2030 emissions
ax1.plot(rcp26_data_m_D2030.Emissions.year[215:336], rcp26_data_m_D2030.Emissions.co2_fossil[215:336], color='blue',label='D2030')
ax2.plot(rcp26_data_m_D2030.Emissions.year[215:336], C26_D2030[:, 0][215:336], color='blue',linestyle='dashed')
ax3.plot(rcp26_data_m_D2030.Emissions.year[215:336], np.sum(F26_D2030, axis=1)[215:336], color='blue',linestyle='dashed')
ax4.plot(rcp26_data_m_D2030.Emissions.year[215:336], T26_D2030[215:336], color='blue',linestyle='dashed')

#this section is my modified RCP26 scenario with the SIAM D2040 emissions
ax1.plot(rcp26_data_m_D2040.Emissions.year[215:336], rcp26_data_m_D2040.Emissions.co2_fossil[215:336], color='red',label='D2040')
ax2.plot(rcp26_data_m_D2040.Emissions.year[215:336], C26_D2040[:, 0][215:336], color='red',linestyle='dashed')
ax3.plot(rcp26_data_m_D2040.Emissions.year[215:336], np.sum(F26_D2040, axis=1)[215:336], color='red',linestyle='dashed')
ax4.plot(rcp26_data_m_D2040.Emissions.year[215:336], T26_D2040[215:336], color='red',linestyle='dashed')

#this section is my modified RCP26 scenario with the SIAM D2050 emissions
ax1.plot(rcp26_data_m_D2050.Emissions.year[215:336], rcp26_data_m_D2050.Emissions.co2_fossil[215:336], color='k',label='D2050')
ax2.plot(rcp26_data_m_D2050.Emissions.year[215:336], C26_D2050[:, 0][215:336], color='k',linestyle='dashed')
ax3.plot(rcp26_data_m_D2050.Emissions.year[215:336], np.sum(F26_D2050, axis=1)[215:336], color='k',linestyle='dashed')
ax4.plot(rcp26_data_m_D2050.Emissions.year[215:336], T26_D2050[215:336], color='k',linestyle='dashed')

#limits
ax1.set_xlim(1980,2100)
ax2.set_xlim(1980,2100)
ax3.set_xlim(1980,2100)
ax4.set_xlim(1980,2100)

#uncertainty lines
#ax3.axvline(x=2100,ymin=0.75,ymax=0.94,color='k',linewidth=4)

#labelling
ax1.set_ylabel('Fossil CO$_2$ Emissions (GtC)')
ax1.legend()
ax2.set_ylabel('CO$_2$ concentrations (ppm)')
ax3.set_ylabel('Total radiative forcing (W m$^{-2}$)')
ax4.set_ylabel('Temperature anomaly (K)');
```
